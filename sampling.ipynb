```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def x(t):
    return np.cos(2*np.pi*0.9*t)

t = np.arange(0,10,0.01)
sinyal = x(t)
plt.plot(t, sinyal)
plt.xlabel("time (second)");
```

```python
freq = np.fft.fft(sinyal)
from matplotlib.colors import LogNorm
plt.plot(np.log10(abs(freq)))
```
