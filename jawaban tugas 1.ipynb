```python
import numpy as np
import scipy.ndimage as ndimage
import matplotlib.pyplot as plt
from skimage import io
from skimage.measure import compare_psnr

%matplotlib inline
```

```python
image = io.imread('lena.jpg')
```

```python
plt.imshow(image, cmap='gray')
```

```python
image
```

```python
lowpass = np.array([[1/9, 1/9, 1/9], [1/9,1/9,1/9], [1/9,1/9,1/9]])
lowpass
```

```python
image_filter = ndimage.convolve(image,lowpass)
image_filter
```

```python
plt.imshow(image_filter, cmap='gray')
```

```python
psnr = compare_psnr(image,image_filter)
psnr
```
