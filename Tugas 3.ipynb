```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

from skimage import color, io
from skimage.transform import resize
%matplotlib inline
```

```python
image_spatial = io.imread("building.jpg")
image_spatial_gray = color.rgb2gray(image_spatial)
```

```python
plt.imshow(image_spatial_gray, cmap='gray')
```

```python
image_spatial_resized = resize(image_spatial_gray,
                               (400,800),
                               order=0,
                               mode='constant',
                               preserve_range=True,
                               anti_aliasing=False)
```

```python
plt.imshow(image_spatial_resized, cmap='gray')
```

```python
image_freq = np.fft.fft2(image_spatial_gray)

from matplotlib.colors import LogNorm
plt.imshow(np.abs(image_freq), norm=LogNorm(vmin=1), cmap='jet')
plt.colorbar()
```

```python
image_freq_downsampled = np.fft.fft2(image_spatial_resized)
plt.imshow(np.abs(image_freq), norm=LogNorm(vmin=1), cmap='jet')
plt.colorbar()
```
