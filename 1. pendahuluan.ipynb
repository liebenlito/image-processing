```python
import numpy as np
import scipy.ndimage as ndimage
import matplotlib.pyplot as plt
from skimage import io

%matplotlib inline
```

```python
image = io.imread('images/cameraman.tif')
```

```python
print(image)
print(image.dtype)
```

```python
plot = plt.imshow(image, cmap='gray')
```

```python
a = np.array([[1,2,0,0],
              [5,3,0,4],
              [0,0,0,7],
              [9,3,0,0]])
```

```python
#ngasal
k = np.array([[-1,9,-1],[-1,9,-1],[-1,9,-1]])
```

```python
l = np.array([[0,-1,0],[-1,2,-1],[0,-1,0]])
```

```python
convl1 = ndimage.convolve(image, k, mode='constant', cval=0.0)
```

```python
convl2 = ndimage.convolve(image, l, mode='constant', cval=0.0)
```

```python
plot = plt.imshow(convl1, cmap='gray')
```

```python
plot = plt.imshow(convl2, cmap='gray')
```

```python
from skimage import filters
```

```python
print(image_sobel)
print(image_sobel.dtype)
```

```python
plot = plt.imshow(image_sobel, cmap='gray')
```
