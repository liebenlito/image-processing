```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import convolve
from skimage import io
%matplotlib inline
```

```python
image_original = io.imread('lena.jpg')
```

```python
plt.imshow(image_original, cmap='gray')
```

```python
#h = [[0,1/6,0],[1/6,1/3,1/6],[0,1/6,0]]
h = [[0,-1/4,0],[-1/4,2,-1/4],[0,-1/4,0]]
```

```python
filter_image = convolve(image_original, h, cval=0.0)
```

```python
plt.imshow(filter_image, cmap='gray')
```
